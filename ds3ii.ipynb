```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Anti.csv")
```

```python
data
```

```python
bins = np.linspace(min(data['V1']), max(data['V1']), 6)
bins
```

```python
group_names = ["b1", "b2", "b3", "b4", "b5"]
```

```python
data['Eqwidth_bin'] = pd.cut(data['V1'], bins, labels = group_names, include_lowest = True)
```

```python
data
```

```python
data['Eqfreq_bins'] = pd.qcut(data['V1'], q=5, labels = ['f1', 'f2', 'f3', 'f4', 'f5'])
```

```python
data
```

```python
condition1 = data['Eqwidth_bin'] == 'b1'
condition2 = data['Eqwidth_bin'] == 'b5'
```

```python
group1 = data[condition1].groupby('Eqwidth_bin')['V1'].transform(list)
```

```python
group2 = data[condition2].groupby('Eqwidth_bin')['V1'].transform(list)
```

```python
data['group1'] = group1
data['group2'] = group2
```

```python
data
```

```python
group1_mean = data['group1'].mean()
group1_median = data['group1'].median()
group1_q1 = data['group1'].quantile(0.25)
group1_q3 = data['group1'].quantile()
group1_std = data['group1'].std()

print(group1_mean)
print(group1_median)
print(group1_q1)
print(group1_q3)
print(group1_std)
```

```python
group2_mean = data['group2'].mean()
group2_median = data['group2'].median()
group2_q1 = data['group2'].quantile(0.25)
group2_q3 = data['group2'].quantile()
group2_std = data['group2'].std()

print(group2_mean)
print(group2_median)
print(group2_q1)
print(group2_q3)
print(group2_std)
```

```python
fig, ax = plt.subplots()
ax.boxplot([data[data['Eqwidth_bin'] == 'b1']['group1']])
ax.set_xticklabels(['b1'])

ax.set_title('Boxplot of group1')
ax.set_xlabel('Eqwidth_bin')
ax.set_ylabel('Virus measure for group1')

plt.show()
```

```python
fig, ax = plt.subplots()
ax.boxplot([data[data['Eqwidth_bin'] == 'b5']['group2']])
ax.set_xticklabels(['b5'])

ax.set_title('Boxplot of group2')
ax.set_xlabel('Eqwidth_bin')
ax.set_ylabel('Virus measure for group2')

plt.show()
```
